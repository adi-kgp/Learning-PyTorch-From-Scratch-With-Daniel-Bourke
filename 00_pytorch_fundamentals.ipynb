### One of the most common errors of deep learning: shape errors

There are two main rules that performing matrix multiplication needs to satisfy:
1. Inner dimensions must match:
* ` (3, 2) @ (3, 2)` wont work
* ` (2, 3) @ (3, 2)` will work

2. The resulting matrix has the shape of the **outer dimensions**
* ` (2,3) @ (3,2) -> (2,2)`
* ` (3,2) @ (2,3) -> (3,3)`

```python
torch.matmul(torch.rand(10,10), torch.rand(10,10)).shape
```

https://www.matrixmultiplication.xyz

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1,2],
                        [3,4],
                        [5,6]])
tensor_B = torch.tensor([[7, 10],
                        [8,11],
                        [9,12]])
```

```python
#torch.mm(tensor_A, tensor_B) # mm alias for matmul
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our tensor shape issues , we can manipulate the shape of one of our tensors using **transpose**


```python
tensor_B.T, tensor_B
```

```python
tensor_B
```

```python
tensor_B.T.shape, tensor_B.shape
```

```python
# The matrix multiplication operations works when tensor_B is transposeed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape}")
output = torch.mm(tensor_A, tensor_B.T)
print(output)
```

## Reshaping, stacking, squeezing and unsqueezing tensors
* Reshaping - reshapes an input tensor to a defined shape
* View - return a view of input tensor of certain shape but keep the same memory of original tensor
* Stacking - combine multiple tensors on top of each other or side by side (h stack)
* Squeeze - removes all '1' dimensions from a tensor
* Unzqueeze - add a `1` dimension to a target tensor
* Permute - return a view of the input with dimensions permuted (swapped) in a certain way

```python
# Let's create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1,9)
z, z.shape
```

```python
# Changing z changes x because the view of a tensor shares the same memory as the original input
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
# torch.squeeze which removes all single dimensions from target tensor
x_reshaped.shape
```

```python
x_reshaped.squeeze()
```

```python
x_reshaped.shape
```

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Removes extra dimensions from x_reshaped
# print(f"\nNew Tensor: {x_squeezed}")
```

```python
# torch.permute rearranges the dimensions of the target tensor in a specified order
x_original = torch.rand(size=(224, 224, 3))
# permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0 ->1, 1->2, 2 ->0
x_permuted.shape, x_original.shape # color channels height, width
```
